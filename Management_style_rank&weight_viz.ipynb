```python
import matplotlib.pyplot as plt

# Sample data (item labels, their corresponding ranks, and weights)
items = ['EMPATHY', 'ENJOYMENT', 'AMIABLE', 'ORDER', 'ANALYTICAL', 'RISKAVERSE', 'AUTHORITATIVE', 'SERIOUS', 'COMPETITIVE', 'RESULTS', 'SPONTANEOUS', 'CURIOUS']
ranks = [10, 8, 6, 5, 4, 3, 3, 2, 2, 1, 1, 1]
weights = [0.2, 0.15, 0.1, 0.08, 0.06, 0.05, 0.04, 0.03, 0.02, 0.02, 0.01, 0.01]

# Calculate weighted ranks
weighted_ranks = [rank * weight for rank, weight in zip(ranks, weights)]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Create a bar plot
bars = ax.barh(items, weighted_ranks, color='skyblue')

# Invert the y-axis to have the highest rank at the top
ax.invert_yaxis()

# Add labels and title
ax.set_xlabel('Weighted Rank')
ax.set_ylabel('Items')
ax.set_title('Weighted Ranked Bar Chart')

# Add weighted ranks as text on the bars
for bar, weighted_rank in zip(bars, weighted_ranks):
    ax.text(bar.get_width() + 0.02, bar.get_y() + bar.get_height()/2, f'{weighted_rank:.2f}', va='center')

plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objects as go

# Sample data (item labels, their corresponding ranks in two groups)
items = ['EMPATHY', 'ENJOYMENT', 'AMIABLE', 'ORDER', 'ANALYTICAL', 'RISKAVERSE', 'AUTHORITATIVE', 'SERIOUS', 'COMPETITIVE', 'RESULTS', 'SPONTANEOUS', 'CURIOUS']
group1_ranks = [10, 8, 6, 5, 4, 3, 3, 2, 2, 1, 1, 1]
group2_ranks = [1, 1, 2, 2, 3, 3, 4, 5, 6, 8, 10, 8]  # Different ranks for group 2

# Create labels for nodes
node_labels = items + items

# Create flows between nodes for group 1
flows_group1 = []
for rank1, rank2 in zip(group1_ranks, group2_ranks):
    flows_group1.append(rank1 - rank2)

# Create flows between nodes for group 2
flows_group2 = []
for rank1, rank2 in zip(group2_ranks, group1_ranks):
    flows_group2.append(rank2 - rank1)

# Create a Sankey chart
fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=node_labels
    ),
    link=dict(
        source=[i for i in range(len(items)) for _ in range(2)],
        target=[len(items) + i for i in range(len(items)) for _ in range(2)],
        value=flows_group1 + flows_group2,
        color=['rgba(0,100,80,0.5)' for _ in range(len(items))] + ['rgba(100,0,80,0.5)' for _ in range(len(items))]
    )
))

# Update layout
fig.update_layout(title_text="Sankey Chart with 12 Items Ranked Differently Between SELF IMAGE and BEHAVIOUR")
fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (item labels, their corresponding ranks in two groups)
items = ['EMPATHY', 'ENJOYMENT', 'AMIABLE', 'ORDER', 'ANALYTICAL', 'RISKAVERSE', 'AUTHORITATIVE', 'SERIOUS', 'COMPETITIVE', 'RESULTS', 'SPONTANEOUS', 'CURIOUS']
self_image_ranks = [10, 8, 6, 5, 4, 3, 3, 2, 2, 1, 1, 1]
behaviour_ranks = [1, 1, 2, 2, 3, 3, 4, 5, 6, 8, 10, 8]

x = np.arange(len(items))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, self_image_ranks, width, label='Self Image', color='skyblue')
rects2 = ax.bar(x + width/2, behaviour_ranks, width, label='Behaviour', color='orange')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Items')
ax.set_ylabel('Ranks')
ax.set_title('Comparison of Ranks Between "Self Image" and "Behaviour"')
ax.set_xticks(x)
ax.set_xticklabels(items, rotation=45, ha="right")
ax.legend()

fig.tight_layout()

plt.show()
```
